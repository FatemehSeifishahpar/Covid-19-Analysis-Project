```python
covid_df.drop(['Sno','Time',  'ConfirmedIndianNational', 'ConfirmedForeignNational'], inplace =True, axis =1)
```

```python
covid_df.head()
```

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'])
```

```python
covid_df.head()
```

```python
# Active cases

covid_df['active_cases'] = covid_df['Confirmed']-(covid_df['Cured']+ covid_df['Deaths'])
```

```python
covid_df.tail(10)
```

```python
statewise = pd.pivot_table(covid_df, values = ['Cured', 'Deaths', 'Confirmed'], index = 'State/UnionTerritory', aggfunc = max)
```

```python
statewise
```

```python
# Recovery Rate
statewise['Recovery_Rate']= statewise['Cured']*100/statewise['Confirmed']
```

```python
# Mortality Rate
statewise['Mortality_Rate']= statewise['Deaths']*100/statewise['Confirmed']
```

```python
statewise = statewise.sort_values(by = 'Confirmed', ascending = False)
```

```python
statewise.style.background_gradient(cmap = 'cubehelix')
```

```python
# Top 10 active cases States:

Top_10_active_cases = covid_df.groupby(by='State/UnionTerritory').max()[['active_cases', 'Date']].sort_values(by = 'active_cases', ascending=False).reset_index()

# Bar plot:

fig = plt.figure(figsize=(10,5))
plt.title("Top 10 active cases", size = 25)
ax = sns.barplot(data = Top_10_active_cases.iloc[:10], y= 'active_cases', x= 'State/UnionTerritory', linewidth=2, edgecolor = 'red')
plt.xlabel("States")
plt.ylabel("Total active cases")
plt.show()
```

```python
Top_10_Deaths = covid_df.groupby(by='State/UnionTerritory').max()[['Deaths', 'Date']].sort_values(by = 'Deaths', ascending=False).reset_index()

# Bar plot:

fig = plt.figure(figsize=(10,5))
plt.title("Top Highest deaths States", size = 25)
ax = sns.barplot(data = Top_10_Deaths.iloc[:10], y= 'Deaths', x= 'State/UnionTerritory', linewidth=2, edgecolor = 'Black')
plt.xlabel("States")
plt.ylabel("Deaths")
plt.show()
```

```python
# Top 5 affected states active cases growth

fig =plt.figure(figsize=(15,9))
ax = sns.lineplot(data=covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra', 'Kamataka', 'Kerala', 'Tamil Nadu', 'Uttar Pradesh'])], x=covid_df['Date'], y=covid_df['active_cases'], hue ='State/UnionTerritory')
ax.set_title('Top 5 affected states in India', size =20)
plt.xlabel("Time")
plt.ylabel("active_cases_growth")
plt.show()
```

```python
# remove rows where state is India

vaccine = vaccine_df[vaccine_df['State']!='India']
```

```python
vaccine.rename(columns ={'Total Individuals Vaccinated': 'Total', }, inplace = True)
vaccine
```

```python
Most_vaccinated= vaccine.groupby(by ='State').sum()[['Total']].sort_values(by='Total', ascending=False)[:7]
Most_vaccinated
```

```python
fig= plt.figure(figsize=(10,5))
ax = sns.barplot(data=Most_vaccinated, y= Most_vaccinated.Total, x= Most_vaccinated.index, linewidth = 2, edgecolor ='red')
plt.xlabel('States')
plt.xticks(rotation=90)
plt.ylabel('Total vaccinated Individuals')
plt.title('Top 7 vaccinated States in India', size=20)
plt.show()
```

```python
Least_vaccinated= vaccine.groupby(by ='State').sum()[['Total']].sort_values(by='Total', ascending=True)[:7]
Least_vaccinated
```

```python
fig= plt.figure(figsize=(10,5))
ax = sns.barplot(data=Least_vaccinated, y= Least_vaccinated.Total, x= Least_vaccinated.index, linewidth = 2, edgecolor ='red')
plt.xlabel('States')
plt.xticks(rotation=90)
plt.ylabel('Total vaccinated Individuals')
plt.title('Leats 7 vaccinated States in India', size=20)
plt.show()
```
